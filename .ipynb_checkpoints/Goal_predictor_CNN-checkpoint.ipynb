```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
raw_data = pd.read_excel("../top5-players24-25.xlsx")
raw_data = raw_data.drop("Rk",axis=1)
df = raw_data[["Pos","Squad","Age","MP","Starts","Min","Gls","Ast","PK","PKatt","xAG"]]
df = df[df["Gls"]>0]
df.dropna(inplace=True)
df
```

```python
categorical = ["Pos","Squad"]
numerical = ["Age","MP","Starts","Min","Gls","Ast","PK","PKatt","xAG"] 
df_encoded = pd.get_dummies(df[categorical + numerical], drop_first=True)
df_encoded.columns
```
